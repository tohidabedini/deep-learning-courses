# Vid 2 ( PDF CDF )

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
from scipy.stats import norm
```

```python
X = np.linspace(-6,6,1000)
```

```python
fx = norm.pdf(X, loc=0, scale=1)
# for norm : loc = mean and scale = standard deviation
plt.plot(X,fx);
# plt.plot(fx);
```

```python
Fx = norm.cdf(X, loc=0, scale=1)
plt.plot(X,Fx);
```

```python
logfx = norm.logpdf(X, loc=0, scale=1)
plt.plot(X,logfx);
```

# Vid 3 ( Convolution )

```python
from PIL import Image
```

```python
im = Image.open('./lenna.png')
gray = np.mean(im, axis=2)
```

```python
x = np.linspace(-6,6,50)
```

```python
fx = norm.pdf(x, loc=0, scale=1)
```

```python
filt = np.outer(fx,fx)
```

```python
plt.imshow(filt,cmap='gray')
```

```python
from scipy.signal import convolve2d
```

```python
out = convolve2d(gray,filt)
```

```python
plt.subplot(1,2,1)
plt.imshow(gray,cmap='gray');
plt.subplot(1,2,2)
plt.imshow(out,cmap='gray');
```

# Vid 4 ( Scipy Exercise )

```python
from PIL import Image
from scipy.signal import convolve2d
im = Image.open('./lenna.png')
gray = np.mean(im, axis=2)
```

```python
Hx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
Hy = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])

Gx = convolve2d(Hx,gray)
Gy = convolve2d(Hy,gray)
```

```python
out = (Gx**2 + Gy**2)**0.5
```

```python
plt.subplot(1,2,1)
plt.imshow(gray,cmap='gray');
plt.subplot(1,2,2)
plt.imshow(out,cmap='gray');
```
