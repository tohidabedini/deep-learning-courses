# Vid 2 ( Line chart )

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,20,100)
y = np.sin(x) + 0.2 * x
```

```python
plt.plot(x,y)
# plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("input")
plt.ylabel("output")
plt.title("my plot");
# semi colon for surpass the last command output
# plt.show()
```

```python
X = np.random.randn(200,2)
X[:50] += 3
Y = np.zeros(200)
Y[:50] = 1
plt.scatter(X[:,0],X[:,1],c=Y)
```

# Vid 6 ( Exercise XOR dataset )

```python
n = 2000
X = 2*np.random.random((n,2)) -1
Y = np.zeros(n)

for i in range(len(X)):
    if (X[i,0] < 0 and X[i,1] > 0 ) or (X[i,0] > 0 and X[i,1] < 0 ):
        Y[i] = 1

plt.scatter(X[:,0],X[:,1],c=Y)
```
